```python
# Referência: https://www.vooo.pro/insights/guia-para-iniciantes-de-web-scraping-em-python-usando-beautifulsoup/
# Importa a biblioteca usada para carregar o website
import urllib.request as urllib2
# Importa as funções de Beautiful Soup que vão parsear os dados obtidos do website
from bs4 import BeautifulSoup
# Importa selenium webdrive
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
# Importa pandas para converter lista em dataframe
import pandas as pd
```

```python
imovel = "https://www.novasaopaulo.com.br/imovel/locacao-casa-americanopolis-sao-paulo/JA18809"
```

```python
#Carrega o website e retorna o html para a variável 'page'
page = urllib2.urlopen(imovel)
```

```python
#Parseia o html na variável 'page', e armazena em formato Beautiful Soup
soup = BeautifulSoup(page)
```

```python
print(soup.prettify())
```
